# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Scatter Plot: Used to visualize the relationship between two numerical variables, showing the distribution of data points.
Line Plot: Used to display the trend of a numerical variable over a continuous interval or time.
Bar Plot: Used to compare categorical data using rectangular bars, where the length of each bar corresponds to the value it represents.
Histogram: Used to visualize the distribution of a single numerical variable by dividing it into bins and showing the frequency of data points in each bin.
Box Plot: Used to show the distribution of data across a categorical variable, displaying key statistics such as the median, quartiles, and potential outliers.

# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. 
Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")

plt.show()
```

# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. 
Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
# Load the dataset
titanic_data = sns.load_dataset("titanic")

# Create box plots
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic_data, x="pclass", y="age")
plt.title("Box Plot of Age by Pclass")

plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic_data, x="pclass", y="fare")
plt.title("Box Plot of Fare by Pclass")

plt.show()
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
# Load the dataset
diamonds_data = sns.load_dataset("diamonds")

# Create a histogram with hue
sns.histplot(data=diamonds_data, x="price", hue="cut", multiple="stack")
plt.title("Histogram of Price by Cut")

plt.show()
```

# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
# Load the dataset
iris_data = sns.load_dataset("iris")

# Create a pair plot with hue
sns.pairplot(data=iris_data, hue="species")
plt.title("Pair Plot of Iris Dataset by Species")

plt.show()
```

# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
# Load the dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix for heatmap
flights_pivot = flights_data.pivot_table(index="month", columns="year", values="passengers")

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data=flights_pivot, cmap="YlGnBu", annot=True, fmt="d")
plt.title("Flights Heatmap")

plt.show()
```
