# Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web framework for Python that is used to build web applications. It is designed to be simple, lightweight, and flexible, providing the essential tools needed to create web applications quickly and efficiently. Flask follows the WSGI (Web Server Gateway Interface) standard and is often used for developing small to medium-sized web applications and APIs. It does not come with built-in features like form validation, database abstraction layers, or other components that are not essential to the core functionality, but these features can be added through extensions.

Advantages of Flask Framework:

Lightweight: Flask is minimalistic and only provides the essentials, allowing developers to choose the components they need.
Flexibility: Developers have the freedom to choose their preferred libraries and tools, making it easy to integrate with other technologies.
Easy to Learn: Flask's simple and intuitive design makes it accessible for beginners, while still offering advanced features for more experienced developers.
Extensible: Flask supports a wide range of extensions that can add functionality such as database integration, form handling, authentication, and more.
Minimal Boilerplate Code: Flask reduces the need for boilerplate code, allowing developers to focus on writing application logic.
Well-documented: Flask has thorough documentation, making it easy to learn and troubleshoot.

# Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello_world():
    return 'Hello World!!'

if __name__ == '__main__':
    app.run()
```

# Q3. What is App routing in Flask? Why do we use app routes?

App routing in Flask refers to the process of mapping URLs to specific functions or views in your application. It allows you to define how different URLs should be handled by your application. Routes are defined using the @app.route() decorator, which associates a URL path with a Python function.

We use app routes in Flask to create different pages or endpoints within our web application. Each route specifies a URL and the corresponding function that should be executed when that URL is accessed. This allows you to create a structured and organized web application with different functionalities accessible via distinct URLs.

# Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999
Attach the screenshot of the output in Jupyter Notebook.

```python
from flask import Flask

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def company_info():
    return '''Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999'''

if __name__ == '__main__':
    app.run()
```

# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

The url_for() function in Flask is used for URL building. It generates a URL for a specific function or endpoint defined in your application, based on its name. This is particularly useful when you want to avoid hardcoding URLs and ensure consistency when generating URLs for different routes.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Home Page'

@app.route('/about')
def about():
    return 'About Us'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('index'))  # Output: /
        print(url_for('about'))  # Output: /about
```
